```python
#BERT 학습용 데이터 넣을 list
disease2 = []
contents2= []

#데이터 베이스 구축용 데이터 프레임 
disease_df = pd.DataFrame(columns=['title','subject','definition', 'cause', 'symptoms', 'diagnosis', 'treatment','progress','prevention','FAQ'])
```

```python
a=0

for link in links_list :
    webpage = requests.get(link)
    soup = BeautifulSoup(webpage.content, "html.parser")

    for i in soup.select('.card-item-inner') : 
        disease = []
        
        webpage = driver.get("http://www.samsunghospital.com/" + i.find("a")["href"])
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")
    
        if  soup.select_one("head > title").get_text() == "ERROR | 삼성서울병원" or (soup.select_one(".post-title > strong").string in (['잠복결핵', '골(퇴행성)관절염','류마티스 관절염', '소아 부정맥','이하선 (침샘 중 귀밑샘) 종양 수술과 안면(얼굴) 신경'])):         
            continue
        else :  
            disease2.append(soup.select_one(".post-title > strong").string)
            contents2.append(soup.select_one(".cms_diseaseDetail"))
            
            disease.append(soup.select_one(".post-title > strong").string)
            
            disease.append(soup.select('div.img')[0].get_text(strip=True))
            if len(soup.select('div.cont')) == 7 :
                for i in range(len(soup.select('div.cont'))):
                    disease.append(soup.select('div.cont')[i].get_text(strip=True))
            else :
                for i in range(len(soup.select('div.cont'))):
                    disease.append(soup.select('div.cont')[i].get_text(strip=True))
                    
                for j in range(7-len(soup.select('div.cont'))) :
                    disease.append(' ')           
            
            disease_df.loc[a]=disease
        
        a = disease_df.index.max()+1
```
